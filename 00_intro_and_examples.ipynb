## 1 — Qu'est-ce que le phishing ?

Le **phishing** (ou hameçonnage) est un type de cyberattaque où un attaquant
envoie un message (généralement un email) qui **prétend provenir d'une source
de confiance** — comme ta banque, ton école, ou un site web populaire — dans
le but de **te piéger** pour te faire faire quelque chose de dangereux :

- Cliquer sur un lien malveillant
- Télécharger un logiciel malveillant (malware)
- Donner ton mot de passe, ton numéro de carte bancaire, ou des informations personnelles

### Phishing vs. Spam

| | Spam | Phishing |
|---|---|---|
| **Objectif** | Faire de la publicité pour un produit ou service | Voler tes données ou installer un malware |
| **Niveau de danger** | Agaçant mais généralement inoffensif | Peut causer de vrais dégâts financiers ou de sécurité |
| **Technique utilisée** | Marketing de masse | Usurpation d'identité d'une entité de confiance |
| **Exemple** | \"Achetez des montres pas chères !\" | \"Votre compte bancaire a été suspendu — cliquez ici pour vérifier\" |

> **Point clé :** tout phishing est un email indésirable, mais tout email indésirable n'est pas du phishing.
> Le phishing cherche spécifiquement à te **tromper** pour te faire effectuer une action nuisible.

## 2 — Signaux d'alerte : comment repérer un email de phishing

L'agence américaine de cybersécurité (CISA) liste plusieurs signes
d'avertissement. Voici les plus courants :

| # | Signal d'alerte | Exemple |
|---|---|---|
| 1 | **Urgence / menaces** | \"Votre compte sera fermé dans 24 heures !\" |
| 2 | **Salutation générique** | \"Cher client\" au lieu de ton nom |
| 3 | **Adresse d'expéditeur suspecte** | `support@paypa1.com` (note le \"1\" au lieu du \"l\") |
| 4 | **Fautes d'orthographe et de grammaire** | \"Nous avons detecté une activité inhabituel sur votre comptes\" |
| 5 | **Liens suspects** | Le texte dit `www.banque.com` mais le vrai lien mène à `site-malveillant.ru` |
| 6 | **Demande d'infos personnelles** | \"Veuillez confirmer votre mot de passe et votre numéro de sécurité sociale\" |
| 7 | **Pièces jointes inattendues** | Un fichier `.zip` ou `.exe` que tu n'as pas demandé |
| 8 | **Trop beau pour être vrai** | \"Vous avez gagné une carte cadeau de 1 000 € !\" |

> Source : [CISA — Reconnaître et éviter les arnaques par email](https://www.cisa.gov/news-events/news/recognizing-and-avoiding-email-scams)

### Lisons quelques emails

Prenons 3 emails sûrs et 3 emails de phishing au hasard et lisons-les.

````python
import textwrap

def show_email(row, max_chars=600):
    """Affiche un email de mani\u00e8re lisible."""
    tag = "PHISHING" if row["label"] == 1 else "S\u00dbR"
    text = row["text"][:max_chars]
    if len(row["text"]) > max_chars:
        text += " [...]"
    wrapped = textwrap.fill(text, width=80)
    print(f"{'='*80}")
    print(f"  Label : {tag}")
    print(f"{'='*80}")
    print(wrapped)
    print()

print("--- 3 emails S\u00dbRS ---\n")
for _, row in df[df["label"] == 0].sample(3, random_state=42).iterrows():
    show_email(row)

print("\n--- 3 emails de PHISHING ---\n")
for _, row in df[df["label"] == 1].sample(3, random_state=42).iterrows():
    show_email(row)
````

## 5 — Pourquoi découper les données ?

En apprentissage automatique (machine learning), on n'évalue jamais un
modèle sur les mêmes données que celles utilisées pour l'entraîner.
Ce serait comme donner les réponses de l'examen à un élève avant le
contrôle — il pourrait tout mémoriser sans rien comprendre.

À la place, on découpe notre jeu de données en **trois parties** :

| Découpage | But | Analogie |
|---|---|---|
| **Ensemble d'entraînement** (~70%) | Le modèle apprend à partir de ces exemples | Le cours et les exercices |
| **Ensemble de validation** (~15%) | On l'utilise pour ajuster les réglages du modèle | Les examens blancs |
| **Ensemble de test** (~15%) | Évaluation finale — utilisé une seule fois à la fin | L'examen final |

Nous avons déjà préparé ces découpages pour toi :

````python
train = pd.read_csv("data/train.csv")
val   = pd.read_csv("data/val.csv")
test  = pd.read_csv("data/test.csv")

print(f"Ensemble d'entra\u00eenement : {len(train):,} emails")
print(f"Ensemble de validation  : {len(val):,} emails")
print(f"Ensemble de test        : {len(test):,} emails")
print(f"{'\u2500'*40}")
print(f"Total                   : {len(train)+len(val)+len(test):,} emails")
````

## 6 — Pourquoi les faux positifs comptent

Quand notre modèle fait une prédiction, il y a quatre résultats possibles :

|  | Le modèle dit : **Sûr** | Le modèle dit : **Phishing** |
|---|---|---|
| **En réalité sûr** | Vrai Négatif (VN) ✅ | **Faux Positif (FP)** ❌ |
| **En réalité phishing** | **Faux Négatif (FN)** ❌ | Vrai Positif (VP) ✅ |

Les deux types d'erreurs sont problématiques, mais de manières différentes :

- **Faux Négatif** (phishing raté) : un email dangereux arrive chez l'utilisateur.
  C'est évidemment mauvais — il pourrait cliquer sur un lien malveillant.

- **Faux Positif** (email sûr signalé comme phishing) : un email légitime est
  bloqué ou envoyé dans le dossier spam. L'utilisateur pourrait rater un message
  important — une offre d'emploi, un rendez-vous médical, ou un message d'un ami.

Dans le monde réel, **les deux erreurs ont un coût**, et il faut trouver le bon
équilibre. Un filtre anti-phishing qui bloque tout est parfaitement sûr mais
complètement inutile.

> **Réfléchis :** Préférerais-tu un filtre qui laisse passer de temps en temps
> un email de phishing, ou un filtre qui bloque 10% de tes emails légitimes ?
> Il n'y a pas de bonne réponse unique — cela dépend du contexte !
